**МАКСИМАЛЬНАЯ ЗАГРУЗКА КЛУБА ПОДОЛЬСК**

```python
#импортируем библиотеки
import pandas as pd
from datetime import datetime, timedelta
```

```python
# импортируем DATAFRAME
pod1= pd.read_excel('D:\idi\работа_про\посещение клубов\Посещения клуба МАЙ ИЮЛЬ 25 ПОДОЛЬСК.xlsx')
pod2= pd.read_excel('D:\idi\работа_про\посещение клубов\Посещения клуба ЯН АПР ПОДОЛЬСК.xlsx')
pod24_1= pd.read_excel('D:\idi\работа_про\посещение клубов\ян-ап24.xlsx')
pod24_2= pd.read_excel('D:\idi\работа_про\посещение клубов\май-авг24.xlsx')
pod24_3= pd.read_excel('D:\idi\работа_про\посещение клубов\сен-дек24.xlsx')
```

```python

# 1. Объединение таблиц
podolsk25=pd.concat([pod1, pod2], ignore_index=True)
podolsk24=pd.concat([pod24_1, pod24_2,pod24_3 ], ignore_index=True)
# 2. Удаление дубликатов (по всем столбцам)
podolsk25 = podolsk25.drop_duplicates()
podolsk24 = podolsk24.drop_duplicates()
#podolsk24=
#приводим Время к типу дататайм
podolsk25['Вход']= pd.to_datetime(podolsk25[' Время'])
podolsk24['Вход']= pd.to_datetime(podolsk24[' Время'])
```

```python
print('Данные по максимальной загрузке клуба ПОДОЛЬСК 2025')
hourly_data = []

for _, row in podolsk25.iterrows():
    current_time = row['Вход']
    end_time = row[' Выход']
    gender = row[' Пол']  # Столбец с полом
    
    # Перебираем каждый час от входа до выхода
    while current_time <= end_time:
        hour_str = current_time.strftime('%Y-%m-%d %H:00:00')
        hourly_data.append({
            'Временной_интервал': hour_str,
            ' Пол': gender,
            'Месяц': current_time.month  # Добавляем месяц для сезонности
        })
        current_time += timedelta(hours=1)

# Создаем DataFrame
hourly_df = pd.DataFrame(hourly_data)
hourly_df['Временной_интервал'] = pd.to_datetime(hourly_df['Временной_интервал'])

# Определяем сезон по месяцу
def get_season(month):
    if 1 <= month <= 4:   # Январь-Апрель
        return 'Январь-Апрель'
    elif 5 <= month <= 8: # Май-Август
        return 'Май-Август'
    else:                 # Сентябрь-Декабрь
        return 'Сентябрь-Декабрь'

hourly_df['Сезон'] = hourly_df['Месяц'].apply(get_season)

# Группируем по времени, полу и сезону
season_gender_counts = hourly_df.groupby(
    ['Сезон', 'Временной_интервал', ' Пол']
).size().unstack(fill_value=0).reset_index()

# Добавляем общую загрузку
season_gender_counts['Общая'] = season_gender_counts.sum(axis=1, numeric_only=True)

# Находим максимальную загрузку для каждого сезона
max_per_season = season_gender_counts.loc[
    season_gender_counts.groupby('Сезон')['Общая'].idxmax()
]

# Выводим результаты
for _, row in max_per_season.iterrows():
    season = row['Сезон']
    time = row['Временной_интервал']
    total = row['Общая']
    
    print(f"\nСезон: {season}")
    print(f"Максимальная загрузка: {total} человек(а)")
    print(f"Дата и час: {time.strftime('%Y-%m-%d %H:00')}")
    
    # Распределение по полу
    print("\nРаспределение по полу:")
    for gender_col in ['М', 'Ж']:
        if gender_col in row:
            print(f"{gender_col}: {row[gender_col]}")
    if 'Другое' in row:
        print(f"Другое: {row['Другое']}")
    
    print("-" * 40)

# Дополнительно: средняя загрузка по сезонам
print("\nСредняя загрузка по сезонам:")
season_avg = season_gender_counts.groupby('Сезон')['Общая'].mean().reset_index()
for _, row in season_avg.iterrows():
    print(f"{row['Сезон']}: {row['Общая']:.2f} человек в час")
```
