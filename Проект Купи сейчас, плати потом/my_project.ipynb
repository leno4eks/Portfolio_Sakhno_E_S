### 3.2. Предобработка данных

```python
# наличие пропусков в данных
if data.isnull().sum().sum()==0:
    print('Датасет  не имеет в себе пропусков')
else:
    print(f'Датасет имеет в себе {data.isnull().sum().sum()} пропусков, необходима очистка ')
    
```

```python
# наличие дубликатов 
if data.duplicated().sum().sum()==0:
    print('Датасет не имеет в себе пропусков')
else:
    print('Датасет имеет в себе дубликаты необходима очистка')
    print(f'Очищаем данные от {data.duplicated().sum().sum()} дубликатов...')
    data= data.drop_duplicates()
    print('Датасет очищен')
```

```python
# наличие выбросов
#  Список числовых колонок 
numeric_cols = data.select_dtypes(include=['int64', 'float64']).columns.tolist()

outlier_stats = []

for i, col in enumerate(numeric_cols):
    # IQR метод
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[col] < lower_bound) | (data[col] > upper_bound)]
    outlier_pct = len(outliers) / len(data) * 100
    outlier_stats.append((col, len(outliers), outlier_pct))

# Вывод статистики по выбросам
print("\nСтатистика выбросов по признакам")
for col, count, pct in outlier_stats:
    print(f"{col:17s}: {count:4d} ({pct:5.2f}%)")
```

```python
#  Признаки из даты
data['transaction_date'] = pd.to_datetime(data['transaction_date'])
data['order_month'] = data['transaction_date'].dt.month
data['order_dayofweek'] = data['transaction_date'].dt.dayofweek  # 0=пн, 6=вс
data['is_weekend'] = (data['order_dayofweek'] >= 5).astype(int)
data['order_year'] = data['transaction_date'].dt.year

data= data.drop('transaction_date', axis=1)
```

```python
# дополниельно, Проверка есть ли повторяющиеся покупки клиентов 
len(data['user_id'].unique())
# удаление признака user_id из-за неинформативности 
data= data.drop('user_id', axis=1)
```

### 4.2. Cоздание новых признаков

Для повышения качества модели необходимо создать новые признаки, отражающие финансовую нагрузку и взаимодействие ключевых факторов. 

`purchase_to_income` – доля стоимости покупки в месячном доходе. Высокое значение означает, что клиент тратит значительную часть заработка на BNPL-транзакцию, что увеличивает риск неплатежа при снижении доходов или непредвиденных расходах.

`install_to_income` – доля одного ежемесячного взноса в доходе. Более детальный показатель, чем общая сумма покупки, так как учитывает количество рассрочек. Чем выше эта доля, тем сильнее платежи давят на бюджет клиента.

`debt_to_credit` – отношение предполагаемого ежемесячного долга (DTI × доход) к кредитному рейтингу. Объединяет информацию о долговой нагрузке и качестве кредитной истории: у клиента с высоким DTI и низким кредитным рейтингом этот показатель резко возрастёт, сигнализируя о крайне высоком 

`delay_times_missed` – произведение дней задержки платежа на число пропусков. Позволяет модели улавливать нелинейное накопление риска: клиент с большим числом пропусков и длительными задержками гораздо опаснее, чем клиент с одним из этих признаков по отдельности.


```python
#  1. Создание новых  признаков для анализа 
# 1.1  Отношение суммы к доходу . Высокое значение может  указывать на риск
data['purchase_to_income']=  data['purchase_amount'] / data['monthly_income']
# 1.2 размер одного платежа к доходу 
data['install_to_income']= (data['purchase_amount'] / data['bnpl_installments']) / data['monthly_income']
# 1.3 Отношение долга к кредитному рейтингу
data['debt_to_credit'] = data['debt_to_income_ratio'] * data['monthly_income'] / data['credit_score'].replace(0, 1)


# 2. Взаимодействия (cross features)
# 2.1. repayment_delay_days * missed_payments
data['delay_times_missed'] = data['repayment_delay_days'] * data['missed_payments']
data['income_to_credit'] = data['monthly_income'] / (data['credit_score'] + 1)

```

#### СТатистические тесты для улавливания корреляции между выбранными категориальными признаками и целевой переменной

1. Критерий χ² (хи-квадрат) для таблиц сопряжённости выбранных призаков и целевой переменной 

Применение: проверка независимости двух признаков.

Нулевая гипотеза H₀: Страна/категория товара  и факт дефолта независимы.

Альтернативная гипотеза H₁: Между страной/категорией товара  и дефолтом существует статистически значимая связь.

```python
# Создаём таблицу сопряжённости
table = pd.crosstab(data['location'], data['default_flag'])

# Выполняем тест
chi2, p, dof, expected = chi2_contingency(table)

print("Хи-квадрат тест: location vs default_flag")
print(f"Статистика χ² = {chi2:.4f}")
print(f"p-value = {p:.4f}")
if p < 0.05:
    print("Вывод: отвергаем H₀ → Между признаком и дефолтом ЕСТЬ статистически значимая связь.")
else:
    print("Вывод: нет оснований отвергнуть H₀ → Связь статистически НЕЗНАЧИМА.")
```

```python
# Создаём таблицу сопряжённости
table = pd.crosstab(data['product_category'], data['default_flag'])

# Выполняем тест
chi2, p, dof, expected = chi2_contingency(table)

print("Хи-квадрат тест: product_category vs default_flag")
print(f"Статистика χ² = {chi2:.4f}")
print(f"p-value = {p:.4f}")

if p < 0.05:
    print("Вывод: отвергаем H₀ → Между признаком и дефолтом ЕСТЬ статистически значимая связь.")
else:
    print("Вывод: нет оснований отвергнуть H₀ → Связь статистически НЕЗНАЧИМА.")
```

```python
from scipy.stats import mannwhitneyu, pointbiserialr

# список числовых признаков (без default_flag)
numeric_features = data.select_dtypes(include=['int64', 'float64']).columns.tolist()
numeric_features.remove('default_flag')  # убираем целевую переменную

results = []

for col in numeric_features:
    # разделяем данные по default_flag
    group0 = data.loc[data['default_flag'] == 0, col].dropna()
    group1 = data.loc[data['default_flag'] == 1, col].dropna()
    
    # U-тест Манна–Уитни
    u_stat, u_p = mannwhitneyu(group0, group1, alternative='two-sided')
    
    # точечно-бисериальная корреляция
    corr, pb_p = pointbiserialr(data['default_flag'], data[col])
    
    results.append({
        'feature': col,
        'u_pvalue': u_p,
        'pb_corr': corr,
        'pb_pvalue': pb_p,
        'significant_u': u_p < 0.05,
        'significant_pb': pb_p < 0.05
    })

# преобразуем в DataFrame для удобного просмотра
res_df = pd.DataFrame(results).sort_values('pb_pvalue')

# выводим признаки, прошедшие оба теста
selected = res_df[(res_df['significant_u']) & (res_df['significant_pb'])]
print("Признаки, статистически значимые по обоим критериям:")
print(selected[['feature', 'u_pvalue', 'pb_corr', 'pb_pvalue']])

# выводим все результаты 
res_df
```

### Построение моделей 

1. Выбор метрики
Поскольку целевая переменная default_flag имеет дисбаланс (≈39% дефолтов), использовать точность (accuracy) нецелесообразно: модель может предсказывать «все не дефолт» и иметь точность 61%, но не выявлять рискованных клиентов.

Основная метрика — `Precision-Recall AUC (PR‑AUC)`.
Она оценивает способность модели находить положительные объекты (дефолты) при заданном уровне точности. В отличие от ROC‑AUC, PR‑AUC чувствительна к дисбалансу классов и отражает реальное качество ранжирования для редкого класса.


Дополнительные метрики (для контроля):

F1-score (гармоническое среднее precision и recall) — хорошо интерпретируется бизнесом.

Recall при фиксированном пороге (например, 0.5) — покажет, какую долю всех дефолтов выявлено.

Precision — доля верно предсказанных дефолтов среди всех предсказанных положительных.



2. Выбор модели

Учитывая табличную природу данных (≈13 000 записей, ≈14 признаков после кодирования), эффективнее всего работают ансамблевые методы, а также логистическая регрессия,  как базовая интерпретируемая модель.

Voting / Stacking:	
- Ансамблирование снижает variance и bias, часто даёт лучшие обобщающие способности.



```python
#  Конфигурация моделей и гиперпараметров

# Общие настройки
CV = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
SCORING = 'average_precision'   # PR-AUC
random_state = 42

# Scale_pos_weight для бустингов
scale_pos = (y_train == 0).sum() / (y_train == 1).sum()

# Словарь: имя -> (модель, param_grid, нужно ли масштабирование)
models_config = {
    'Logistic Regression': (
        LogisticRegression(class_weight='balanced', random_state=random_state, max_iter=1000),
        {'C': [0.01, 0.1, 1, 20], 'penalty': ['l2', 'l1'], 'solver': ['liblinear']},
        True   # нужен StandardScaler
    ),
    'Random Forest': (
        RandomForestClassifier(class_weight='balanced', random_state=random_state, n_jobs=-1),
        {'n_estimators': [100, 300], 'max_depth': [5, 10, None], 'min_samples_split': [2, 5]},
        False
    ),
    'XGBoost': (
        xgb.XGBClassifier(scale_pos_weight=scale_pos, random_state=random_state,
                          use_label_encoder=False, eval_metric='logloss'),
        {'n_estimators': [100, 300], 'max_depth': [3, 5, 7],
         'learning_rate': [0.01, 0.05, 0.1], 'subsample': [0.8, 1.0],
         'colsample_bytree': [0.8, 1.0]},
        False
    ),
    'LightGBM': (
        lgb.LGBMClassifier(class_weight='balanced', random_state=random_state, verbose=-1),
        {'n_estimators': [100, 300], 'max_depth': [3, 5, 7],
         'learning_rate': [0.01, 0.05, 0.1], 'subsample': [0.8, 1.0],
         'colsample_bytree': [0.8, 1.0]},
        False
    ),
    'CatBoost': (
        CatBoostClassifier(class_weights=[1, scale_pos], random_seed=random_state, verbose=0),
        {'depth': [3, 5, 7], 'learning_rate': [0.01, 0.05, 0.1],
         'l2_leaf_reg': [1, 3, 5], 'iterations': [100, 300]},
        False
    )
}
```

```python

# Функция обучения и оценки  модели
def train_and_evaluate(name, model, param_grid, use_scaler, X_train, y_train, X_test, y_test, cv=CV, scoring=SCORING):
    """
    Обучает модель с GridSearchCV, оценивает на тесте.
    Возвращает словарь с метриками и обученный классификатор.
    """
    if use_scaler:
        pipe = Pipeline([('scaler', StandardScaler()), ('clf', model)])
        param_grid = {f'clf__{k}': v for k, v in param_grid.items()}
    else:
        pipe = model

    grid = GridSearchCV(pipe, param_grid, cv=cv, scoring=scoring, n_jobs=-1)
    grid.fit(X_train, y_train)

    best_model = grid.best_estimator_
    # Для получения вероятностей нужно извлекать классификатор из Pipeline
    if use_scaler:
        y_proba = best_model.predict_proba(X_test)[:, 1]
        y_pred = (y_proba >= 0.5).astype(int)
    else:
        y_proba = best_model.predict_proba(X_test)[:, 1]
        y_pred = best_model.predict(X_test)

    metrics = {
        'Model': name,
        'PR-AUC': average_precision_score(y_test, y_proba),
        'F1': f1_score(y_test, y_pred),
        'Recall': recall_score(y_test, y_pred),
        'Precision': precision_score(y_test, y_pred),
        'Best Params': grid.best_params_,
        'Model Object': best_model
    }
    return metrics
```

```python

# Обучение всех моделей и сбор результатов

results = []
trained_models = {}   # для последующего ансамблирования

for name, (model, param_grid, use_scaler) in models_config.items():
    print(f"\n--- Тренировка {name} ---")
    metrics = train_and_evaluate(name, model, param_grid, use_scaler,
                                 X_train, y_train, X_test, y_test)
    results.append({k: v for k, v in metrics.items() if k not in ['Model Object', 'Best Params']})
    trained_models[name] = metrics['Model Object']
    print(f"PR-AUC: {metrics['PR-AUC']:.4f} | F1: {metrics['F1']:.4f}")
```

```python

# Отдельно делаем модели для ансамблей: Voting и Stacking

# Voting 
estimators = [(name.lower().replace(' ', '_'), model) for name, model in trained_models.items()]
voting = VotingClassifier(estimators, voting='soft')
voting.fit(X_train, y_train)
y_proba_vote = voting.predict_proba(X_test)[:, 1]
y_pred_vote = (y_proba_vote >= 0.5).astype(int)

voting_metrics = {
    'Model': 'Voting (soft)',
    'PR-AUC': average_precision_score(y_test, y_proba_vote),
    'F1': f1_score(y_test, y_pred_vote),
    'Recall': recall_score(y_test, y_pred_vote),
    'Precision': precision_score(y_test, y_pred_vote)
}
results.append(voting_metrics)

# Stacking
meta_model = LogisticRegression(class_weight='balanced', max_iter=1000)
stacking = StackingClassifier(estimators=estimators, final_estimator=meta_model, cv=5)
stacking.fit(X_train, y_train)
y_proba_stack = stacking.predict_proba(X_test)[:, 1]
y_pred_stack = (y_proba_stack >= 0.5).astype(int)

stack_metrics = {
    'Model': 'Stacking',
    'PR-AUC': average_precision_score(y_test, y_proba_stack),
    'F1': f1_score(y_test, y_pred_stack),
    'Recall': recall_score(y_test, y_pred_stack),
    'Precision': precision_score(y_test, y_pred_stack)
}
results.append(stack_metrics)

# Stacking с оптимизацией порога
prec, rec, thresholds = precision_recall_curve(y_test, y_proba_stack)
f1_scores = 2 * (prec * rec) / (prec + rec + 1e-9)
best_idx = np.argmax(f1_scores[:-1])
best_thr = thresholds[best_idx]
y_pred_stack_opt = (y_proba_stack >= best_thr).astype(int)

stack_opt_metrics = {
    'Model': 'Stacking (opt)',
    'PR-AUC': average_precision_score(y_test, y_proba_stack),
    'F1': f1_score(y_test, y_pred_stack_opt),
    'Recall': recall_score(y_test, y_pred_stack_opt),
    'Precision': precision_score(y_test, y_pred_stack_opt)
}
results.append(stack_opt_metrics)
```

```python

# Итоговая таблица результатов

results_df = pd.DataFrame(results)
print("\n" + "="*60)
print("Сравнение моделей")
print("="*60)
print(results_df.round(4).to_string(index=False))
```
